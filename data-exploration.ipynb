```python
try:
    import kagglehub
except ModuleNotFoundError:
    print("The 'kagglehub' module is not installed. Please install it using pip: 'pip install kagglehub'")

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(path + "/books.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False , ax = ax)
plt.xlabel('columns')
plt.ylabel('missing')
plt.show()
```

```python
import numpy as np
books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['age_of_book'] = 2025 - books['published_year']
```

```python
columns_of_interest = ['num_pages', 'age_of_book', 'missing_description', 'average_rating']
correlation_matrix = books[columns_of_interest].corr(method = 'spearman') # spearman correlation for non continuous data and pearson correlation for continuous data
# spearman correlation for non-continuous data because of the missing values .
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(correlation_matrix , annot=True, cmap='coolwarm', fmt=".2f",cbar_kws={'label': 'Spearman Correlation'})
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
plt.show()
```
